```python
#Measuring Accuracy
```

```python
from sklearn.metrics import mean_squared_error
```

```python
Grades_prediction = lin_reg.predict(features_std)
```

```python
import numpy as np
lin_mse = mean_squared_error(grade_label, Grades_prediction)
lin_rmse = np.sqrt(lin_mse)
```

```python
lin_rmse
```

```python
#2nd Algorithm and Error Check
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(features_std, grade_label)
grade_pred = tree_reg.predict(features_std)
tree_mse = mean_squared_error(grade_label, grade_pred)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

```python
#3rd Algorithm and error check
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(features_std, grade_label)
grade_pred = forest_reg.predict(features_std)
forest_mse = mean_squared_error(grade_label, grade_pred)
forest_rmse = np.sqrt(tree_mse)
forest_rmse
```

```python
#Validation for Random Forest Regression
from sklearn.model_selection import cross_val_score
scores = cross_val_score(forest_reg, features_std, grade_label, scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-scores)
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation", scores.std())
display_scores(forest_rmse_scores)
```

```python
#Validation for Linear Regression
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lin_reg, features_std, grade_label, scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-scores)
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation", scores.std())
display_scores(lin_rmse_scores)
```

```python
#Validation for Decision Tree Regression
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, features_std, grade_label, scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation", scores.std())
display_scores(tree_rmse_scores)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = [
    {'n_estimators': range(1, 40), 'max_features': range(1, 3), 'max_depth': range(1, 5)}, 
    {'bootstrap': [False], 'n_estimators': range(1, 40), 'max_features': range(1, 3), 'max_depth': range(1, 5)}
]
forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg, param_grid, cv=10, scoring='neg_mean_squared_error')
grid_search.fit(features_std, grade_label)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
final_model = grid_search.best_estimator_
final_predictions = final_model.predict(features_std)
final_mse = mean_squared_error(grade_label, final_predictions)
final_rmse = np.sqrt(final_mse)
final_rmse
```
